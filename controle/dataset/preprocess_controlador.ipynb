# Imports

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

# Helper functions

# Preprocess

```python
df_raw_controle = pd.read_csv('/content/dataset_controle_raw.csv')
df_raw_controle = df_raw_controle[['expected_ul', 'expected_ur', 'real_ul', 'real_ur']]

```

```python
for col in df_raw_controle.columns:
    df_raw_controle[col] = pd.to_numeric(df_raw_controle[col], errors='coerce')

df_raw_controle.dropna(inplace=True)

df_raw_controle = df_raw_controle.astype(float)
```

```python

df_raw_controle.head(5)
```

```python
df_preprocessed_controle = df_raw_controle[['expected_ul', 'expected_ur']].copy()

df_preprocessed_controle['real_ul_m1'] = df_raw_controle['real_ul'].shift(1).fillna(0)
df_preprocessed_controle['real_ur_m1'] = df_raw_controle['real_ur'].shift(1).fillna(0)

df_preprocessed_controle['delta_ul'] =  df_raw_controle['expected_ul'] - df_raw_controle['real_ul']
df_preprocessed_controle['delta_ur'] =  df_raw_controle['expected_ur'] - df_raw_controle['real_ur']

df_preprocessed_controle
```

```python
df_train, df_test = train_test_split(df_preprocessed_controle, test_size=0.2, random_state=2024)
```

```python

df_train.to_csv('/content/to_train.csv')
df_test.to_csv('/content/to_test.csv')
```
