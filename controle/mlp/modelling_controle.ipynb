# Imports

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt


import numpy as np
import pandas as pd

from joblib import dump
```

# Helper functions

# Loading dataframe

```python
df_train = pd.read_csv('/content/to_train (1).csv')
df_test = pd.read_csv('/content/to_test (2).csv')

df_train = df_train.drop('Unnamed: 0', axis=1)
df_test = df_test.drop('Unnamed: 0', axis=1)
```

```python
df_test.head(1)
```

```python

df_train.columns
```

# Modelling

```python
for col in ['expected_ul', 'expected_ur', 'real_ul_m1', 'real_ur_m1', 'delta_ul', 'delta_ur']:
    if not np.issubdtype(df_train[col].dtype, np.number):
        df_train = df_train[pd.to_numeric(df_train[col], errors='coerce').notnull()]
        df_test = df_test[pd.to_numeric(df_test[col], errors='coerce').notnull()]

X_train = df_train[['expected_ul', 'expected_ur', 'real_ul_m1', 'real_ur_m1']]
y_train = df_train[['delta_ul', 'delta_ur']]
X_test = df_test[['expected_ul', 'expected_ur', 'real_ul_m1', 'real_ur_m1']]
y_test = df_test[['delta_ul', 'delta_ur']]
```

```python
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50,50), (100,50)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
}

mlp = MLPRegressor(max_iter=500, verbose=True, early_stopping=True)

grid_search = GridSearchCV(mlp, param_grid, cv=10, scoring='neg_mean_squared_error', verbose=3)

grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print(f'RMSE: {rmse}\nMSE: {mse}\nMAE: {mae}')

diff_abs = np.abs(y_test - y_pred)
limite = 0.3 * np.abs(y_test)
acertos = np.sum((diff_abs <= limite), axis=0)
acuracia = acertos / len(y_test) * 100
print(f'AcurÃ¡cia +/- 30%: {acuracia}%')

plt.plot(best_model.loss_curve_)
plt.title('Loss Curve')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.show()
```

# Saving best model

```python
dump(best_model, 'control_model.pkl')
```
